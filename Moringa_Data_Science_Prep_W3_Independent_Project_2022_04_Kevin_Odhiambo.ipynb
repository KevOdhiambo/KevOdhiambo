```python
#chnage CDR columns to match for easier merging  
data4.columns = map(str.lower, data4.columns)
data4.columns
#assign the columns titles in lowercase
columns = ['product', 'value', 'date_time', 'cell_on_site', 'dw_a_number_int',
'dw_b_number_int', 'country_a', 'country_b', 'cell_id', 'site_id']


```

```python
#change data3 columns to lower case for easier merging
data3.columns=map(str.lower, data3.columns)
data3.columns

data3.head()
```

```python
#chnage data4 columns to lower case and preview columns
data4.columns=columns
data4.head()
```

```python
#change data5 columns to lower case and preview columns
data5.columns=columns
data5.head()
```

```python
#change data6 columns to lower case and preview columns
data6.columns=columns
data6.head()
```

```python
#Concatinating the CDR datasets to work on one single table
data7=pd.concat([data4,data5,data6], axis=0)
data7.info()
```

```python
#drop Country A and B, as they are not needed
data7.drop(['country_a', 'country_b'], axis=1)
```

```python
#check and remove duplicates from the data
data7 = data7.drop_duplicates()
data7.info()
#more than 250 duolicates have been removed
```

```python
#change the date type in the columns to date type 
data7['date_time'] = pd.to_datetime(data7['date_time'])


#preview the data type for date_time
data7.info()
```

```python
#Join Data3 and Data7 for easier data manipulation

data8 = pd.merge(data3,data7, how='inner', on=['cell_id'])
data8.info()
```

```python
#Which ones were the most used city for the three days?
# here we count calls per city, and one with the most calls is the most used city. 

data7.groupby('cell_id').count().nlargest(10,'product')

```

```python
#getting the cell with the highest billing
data7.groupby('site_id')['value'].max()
```
